### 1. Reconstruct test audio

```python
def reconstruct_sound(signal, config):
    device = (
        torch.device("cuda")
        if torch.cuda.is_available()
        else torch.device("mps")
        if torch.backends.mps.is_available()
        else torch.device("cpu")
    )
    print(device)

    model, _, _ = make_model(config, device, train=False)
    out, _, _ = model(signal.reshape(1, 1, signal.shape[1]).to(device))

    return out.cpu().numpy().reshape(1, out.shape[2])
```

```python
processor = DataProcessor(config["dataset"])
dataset = processor.test_dataset
idx = random.randint(0, len(dataset))
signal = dataset[idx]
Audio(signal, rate=44_100)
```

```python
new_signal = reconstruct_sound(signal, config)
```

```python
Audio(new_signal, rate=44_100)
```

### 2. Generating new sounds

```python
def generate_sound(config):
    device = (
        torch.device("cuda")
        if torch.cuda.is_available()
        else torch.device("mps")
        if torch.backends.mps.is_available()
        else torch.device("cpu")
    )
    model, _, _ = make_model(config, device, train=False)
    out = model.generate()

    return out.cpu().numpy().reshape(1, out.shape[2])
```

```python
gen_sound = generate_sound(config)
Audio(gen_sound, rate=44_100)
```
